```python
import pandas as pd
import string
```

```python
data = pd.read_csv("passwords.csv")
data
```

```python
data =  data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
data['password']=data['password'].astype('str')
```

```python
data['length'] = data['password'].str.len()
```

```python
def count_capital_letters(password):
    return sum(1 for char in password if char.isupper())

def count_small_letters(password):
    return sum(1 for char in password if char.islower())

def count_numbers(password):
    return sum(1 for char in password if char.isdigit())

def count_duplicate_characters(password):
    duplicate_count = 0
    password = str(password)
    password_length = len(password)
  
    for i in range(password_length):
        for j in range(i + 1, password_length):
            if password[i] == password[j]:
                duplicate_count += 1
                break
  
    return duplicate_count

def count_special_characters(password):
    special_chars = string.punctuation
    return sum(1 for char in password if char in special_chars)

data['uppercase'] = data['password'].apply(count_capital_letters)
data['lowercase'] = data['password'].apply(count_small_letters)
data['numbers'] = data['password'].apply(count_numbers)
data['duplicate'] = data['password'].apply(count_duplicate_characters)
data['special_characters'] = data['password'].apply(count_special_characters)
```

```python
data.to_csv('passwords_final.csv',index=None)
```
